```python
#Import required libraries
from sklearn.datasets import load_iris
import pandas as pd

#Load the Iris dataset
iris = load_iris()

#Convert dataset into a Pandas DataFrame with feature names as columns
df = pd.DataFrame(iris.data, columns=iris.feature_names)

#Add species column by converting numeric target values into categorical class names
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

#Import matplotlib for visualization
import matplotlib.pyplot as plt

#Define colors for each species for better visualization
colors = {'setosa': 'red', 'versicolor' : 'blue', 'virginica' : 'green'}

#Create scatter plot for Sepal Length vs Sepal width
plt.scatter(df['sepal length (cm)'],
            df['sepal width (cm)'],
            c=df['species'].map(colors))

#Label x-axis
plt.xlabel("Sepal Length")

#Label y-axis
plt.ylabel("Sepal Width")

#Add title to the plot
plt.title("Scatter Plot of Iris Dataset")

#Display the plot
plt.show()
```

**Scatter plot** of sepal length versus sepal width shows that each point on the graph represents an individual flower, with its sepal length on the x-axis and sepal width on the y-axis. The different colors distinguish between the three species of iris flowers: setosa (red), versicolor (blue) and virginica (green). This visualization helps in observing the distribution and separation of these species based on their sepal dimensions. The data points are moderately spread indicating a weak to moderate relationship between these two features. The different Iris species show some overlap, meaning they are not very clearly separable using only sepal measurements.

```python
#Create scatter plot for Petal Length vs Petal Width
plt.scatter(df['petal length (cm)'],
            df['petal width (cm)'],
            c=df['species'].map(colors))

#Label x-axis
plt.xlabel("Petal Length")

#Label y-axis
plt.ylabel("Petal Width")

#Add title to the plot
plt.title("Scatter Plot of Iris Dataset (Petal Dimensions)")

#Display the plot
plt.show()
```

**Scatter plot** visualizes the relationship between petal length and petal width for the Iris dataset. Similar to the sepal plot, each point represents a flower, with its petal length on the x-axis and petal width on the y-axis. The colors indicate the different iris species, allowing us to observe how these dimensions contribute to the classification of each species. It shows a strong positive relationship as the points form clearer clusters and different Iris species are more easily separable compared to sepal features. This indicates that petal measurements are more useful for distinguishing between species.

```python
#Import matplotlib library for plotting graphs
import matplotlib.pyplot as plt

#Create histograms for all numerical columns in the dataset
#figsize sets the size of the figure (width=10, height=8)
#bins=10 divides the data into 10 intervals for better visualization of the distribution
df.hist(figsize=(10,8), bins=10)

#Display the plot
plt.show()
```

**Histogram** is a type of graph used to understand how data is distributed. In this graph, figsize=(10,8) is used to set the size of the plot where 10 represents the width and 8 represents the height, making the graph clearer and easier to read. The parameter bins=10 divides the data into 10 equal intervals or groups, which helps in showing how frequently value falls within specific ranges. By adjusting the number of bins, we can see the data in more detailed or more general form depending on analysis requirement.

```python
#Import matplotlib library for data visualization
import matplotlib.pyplot as plt

#Set the size of the figure (width=10, height=6) for better readability
plt.figure(figsize=(10,6))

# Select only numerical columns by removing the categorical 'species' column
numerical_df = df.drop(columns=['species'])

#Create a box plot to visualize the distribution, spread and outliers of numerical features
plt.boxplot(numerical_df.values)

#Add title to the plot
plt.title("Box Plot of Iris Features")

#Set x-axis labels to match column names and rotate them for better readability
plt.xticks([1,2,3,4], numerical_df.columns, rotation=45)

#Label the y-axis
plt.ylabel("Values")

#Display the plot
plt.show()
```

**Box plot** displays the distribution of each numerical feature (sepal length, sepal width, petal length and petal width) showing the potential outliners. This visualization provides a quick summary of the data spread and central tendency for each measurement.

**Exploratory Data Analysis (EDA)** is the process of analyzing a dataset visually before applying any machine learning model or drawing conclusions. It helps in understanding the structure, patterns, trends and relationships within the data using different types of visualizations such as histograms, scatter plots, box plots and bar charts. EDA is important because it allows us to detect missing values, outliners and hidden patterns making it easier to decide which features are useful and how the data should be prepared for further analysis.

**Conslusion** The Exploratory Data Analysis (EDA) of the Iris Dataset helped in understanding the distribution, relationships and variability of the features. Histogram showed how the data is spread, scatter plots revealed patterns between variables and box plots identified outliners and differences in future spread. Overall, the analysis confirmed that the dataset contains clear patterns suitable for further machine learning tasks.
